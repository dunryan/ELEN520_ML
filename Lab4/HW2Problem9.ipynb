```python
# What is the mean and standard deviation of each quantitative predictor?
import numpy as np
print('Range of mpg values is: ',np.ptp(df['mpg']))
print('Range of cylinder values is: ',np.ptp(df['cylinders']))
print('Range of displacements is: ',np.ptp(df['displacement']))
print('Range of weight values is: ',np.ptp(df['weight']))
print('Range of acceleration values is: ',np.ptp(df['acceleration']))
#print('Range of horsepower values is: ',np.ptp(df['horsepower']))
```

```python
print('Standard deviation of mpg values is: ',np.std(df['mpg']))
print('Standard deviation of cylinder values is: ',np.std(df['cylinders']))
print('Standard deviation of displacements is: ',np.std(df['displacement']))
print('Standard deviation of weight values is: ',np.std(df['weight']))
print('Standard deviation of acceleration values is: ',np.std(df['acceleration']))
```

```python
print('Mean of mpg values is: ',np.mean(df['mpg']))
print('Mean of cylinder values is: ',np.mean(df['cylinders']))
print('Mean of displacements is: ',np.mean(df['displacement']))
print('Mean of weight values is: ',np.mean(df['weight']))
print('Mean of acceleration values is: ',np.mean(df['acceleration']))
```
