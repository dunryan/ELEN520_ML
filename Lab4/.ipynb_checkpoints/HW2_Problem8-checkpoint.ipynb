```python
# scatter plot matrix of first ten columns
import seaborn as sns
sns.pairplot(df.iloc[:, 1:11])
```

```python
#Create a new qualitative variable, called Elite, by binning
#the Top10perc variable. We are going to divide universities
#into two group based on whether or not the proportion
#of students coming from the top 10% of their high school
#classes exceed 50 %.

df['Elite'] = df['Top10perc'] >50
df['Elite'].sum()
```

```python
sns.boxplot(x=df['Elite'],y=df['Outstate'])
```
